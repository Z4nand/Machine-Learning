## Анализ данных и препросессинг

```python
df_eye=df['eyesight(left, right)'].str.split(',',expand=True) #Разбиение колонки на 2
df_eye.columns = ['eyesight_left', 'eyesight_right']
```

```python
df_hear = df['hearing(left, right)'].str.split(',',expand=True) #Аналогичное разбиение на 2
df_hear.columns = ['hear_left', 'hear_right']
```

```python
df = df.join(df_eye) #добавляем перекодированные столбцы
df = df.join(df_hear)
```

```python
df=df.drop(columns = ['eyesight(left, right)', 'hearing(left, right)']) #удаление непереработанных столбцов
```

```python
df.head()
```

```python
mapping = {'Y': 1,
           'N': 0}

df['dental caries'] = df['dental caries'].map(mapping)
df['tartar'] = df['tartar'].map(mapping)
df['smoking'] = df['smoking'].map(mapping)

df['gender'] = df['gender'].map({'M': 0, 'F': 1})


```

```python
df.head()
```

```python
df['age'].notna().sum()
```

```python
print(df.notna().sum())
df = df.fillna(method='pad')
df = df.fillna(method='backfill')
```

```python
df.tail()
```

### Отбор на основе важности

```python
from sklearn.ensemble import ExtraTreesClassifier 

X = df.iloc[:, :-1].values
y = df["smoking"]
# feature extraction
model = ExtraTreesClassifier()
model.fit(X, y)
print(model.feature_importances_)
```
